# Seawater Velocity from Axial Seamount Shallow Profiler in 2018

# Import Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import glob
```

# Calculate Absolute Value of Eastward Seawater Velocity

```python
filenames = sorted(glob.glob('*.csv'))
filenames = filenames[0:182]
avgVelocity = []
for f in filenames:
    data = np.genfromtxt(fname=f, delimiter=',', skip_header=1) 
    avgVelocity.append(np.mean(abs(data[:,25])));
```

# Plot Average Seawater Velocity in each Month

```python
jan = np.arange(0,29,1);
feb = np.arange(29,57,1);
march = np.arange(58,88,1);
april = np.arange(89,118,1);
may = np.arange(119,149,1);
june = np.arange(150,179,1);
july = np.arange(180,182,1);
#aug = np.arange(208,238,1);
#sept = np.arange(239,268,1);
#octo = np.arange(269,299,1);
#nov = np.arange(300,329,1);
#dec = np.arange(330,363,1);

day = np.arange(0,182,1)
plt.plot(day,avgVelocity,'w--')
plt.plot(jan,avgVelocity[0:29],'darkturquoise',markersize = 3)
plt.plot(feb,avgVelocity[29:57],'royalblue',markersize = 3)
plt.plot(march,avgVelocity[58:88],'limegreen',markersize = 3)
plt.plot(april,avgVelocity[89:118],'olivedrab',markersize = 3)
plt.plot(may,avgVelocity[119:149],'yellowgreen',markersize = 3)
plt.plot(june,avgVelocity[150:179],'orangered',markersize = 3)
plt.plot(july,avgVelocity[180:182],'gold',markersize = 3)
#plt.plot(aug,avgVelocity[208:238],'darkorange',markersize = 3)
#plt.plot(sept,avgVelocity[239:268],'saddlebrown',markersize = 3)
#plt.plot(octo,avgVelocity[269:299],'chocolate',markersize = 3)
#plt.plot(nov,avgVelocity[300:329],'maroon',markersize = 3)
#plt.plot(dec,avgVelocity[330:363],'dodgerblue',markersize = 3)

plt.title('Eastward Seawater Velocity: Axial Seamount Shallow Profiler');
plt.xlabel("Day of Year: Jan 1 - July 2");
plt.ylabel("Average Seawater Velocity [m s-1]")
plt.grid(True)
```
