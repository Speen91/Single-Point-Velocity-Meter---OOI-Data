# Seawater Velocity from Axial Seamount Shallow Profiler in 2017

# Import Packages

```python
import numpy as np
import matplotlib.pyplot as plt
import glob
```

# Calculate Absolute Value of Eastward Seawater Velocity

```python
filenames = sorted(glob.glob('*.csv'))
filenames = filenames[0:361]
avgVelocity = []
for f in filenames:
    data = np.genfromtxt(fname=f, delimiter=',', skip_header=1) 
    avgVelocity.append(np.mean(abs(data[:,25])));
```

# Plot Average Seawater Velocity in each Month

```python
jan = np.arange(0,29,1);
feb = np.arange(29,57,1);
march = np.arange(58,88,1);
april = np.arange(89,118,1);
may = np.arange(119,149,1);
june = np.arange(150,179,1);
july = np.arange(180,210,1);
aug = np.arange(211,239,1);
sept = np.arange(240,269,1);
octo = np.arange(270,300,1);
nov = np.arange(301,330,1);
dec = np.arange(331,361,1);

day = np.arange(0,361,1)
plt.plot(day,avgVelocity,'k--')
plt.plot(jan,avgVelocity[0:29],'darkturquoise',markersize = 3)
plt.plot(feb,avgVelocity[29:57],'royalblue',markersize = 3)
plt.plot(march,avgVelocity[58:88],'limegreen',markersize = 3)
plt.plot(april,avgVelocity[89:118],'olivedrab',markersize = 3)
plt.plot(may,avgVelocity[119:149],'forestgreen',markersize = 3)
plt.plot(june,avgVelocity[150:179],'orangered',markersize = 3)
plt.plot(july,avgVelocity[180:210],'darkorange',markersize = 3)
plt.plot(aug,avgVelocity[211:239],'crimson',markersize = 3)
plt.plot(sept,avgVelocity[240:269],'saddlebrown',markersize = 3)
plt.plot(octo,avgVelocity[270:300],'tan',markersize = 3)
plt.plot(nov,avgVelocity[301:330],'chocolate',markersize = 3)
plt.plot(dec,avgVelocity[331:361],'dodgerblue',markersize = 3)

plt.title('Eastward Seawater Velocity: Axial Seamount Shallow Profiler');
plt.xlabel("Day of Year: Jan 1 - Dec 31");
plt.ylabel("Average Seawater Velocity [m s-1]")
plt.grid(True)
```
